```python
### Begin Assignment Here

# DM2025 - Lab1 Phase 1 on Reddit-stock-sentiment (All-in-One Cell, English-only)
# - Mirrors Exercises 1–15 functionality from the master, but uses the Reddit dataset instead of 20newsgroups.
# - Clean outputs, no seaborn/plotly, matplotlib only.

import sys, re, html
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
import joblib
import types
import warnings

# ---- Display & warnings (tidy outputs) ----
pd.set_option("display.max_rows", 50)
pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 120)
pd.set_option("display.max_colwidth", 120)
warnings.filterwarnings("ignore", category=FutureWarning)

print("Python:", sys.version.split()[0])

# ==========================================================
# 0) Load and prepare the dataset (keep only text + label)
# ==========================================================
URL = "https://raw.githubusercontent.com/leoson-wu/DM2025-Lab1-Exercise/main/newdataset/Reddit-stock-sentiment.csv"
raw = pd.read_csv(URL)

label_map = {-1.0: "negative", 0.0: "neutral", 1.0: "positive"}
X = (
    raw.loc[:, ["text", "label"]]
       .dropna(subset=["text", "label"])
       .assign(label=lambda d: d["label"].astype(float).clip(-1, 1))
)
X = X[X["label"].isin([-1.0, 0.0, 1.0])].copy()
X["category_name"] = X["label"].map(label_map).astype("category")
X["text"] = X["text"].astype(str)
X.reset_index(drop=True, inplace=True)

print("Dataset shape:", X.shape)
print("Label counts:\n", X["category_name"].value_counts())

# --- X_sample: stratified sample (to mimic master’s smaller sample) ---
def stratified_sample(df: pd.DataFrame, label_col: str, frac: float = 0.2, random_state: int = 42) -> pd.DataFrame:
    """
    Stratified sampling by a categorical label column using GroupBy.sample.
    - observed=True for clean grouping behavior on categoricals
    - group_keys=False to avoid MultiIndex concat artifacts
    - Guarantees at least 1 sample per class
    """
    rng = np.random.default_rng(random_state)
    parts = []
    for name, sub in df.groupby(label_col, observed=True, group_keys=False):
        n = max(1, int(round(len(sub) * frac)))
        seed = int(rng.integers(0, 2**31 - 1))  # deterministic per class, but distinct
        parts.append(sub.sample(n=n, random_state=seed))
    return pd.concat(parts, ignore_index=True)

X_sample = stratified_sample(X, "category_name", frac=0.2, random_state=42)

# ==========================================================
# 1) Print first 3 documents (mirror: twenty_train.data[:3])
# ==========================================================
twenty_train = types.SimpleNamespace(data=X["text"].tolist())
print("\n# First 3 documents:")
for i in range(min(3, len(twenty_train.data))):
    print(f"\n--- DOC {i} ---\n{twenty_train.data[i][:1000]}")  # truncated for readability

# ==========================================================
# 2) Pandas querying techniques on X (mirrors master behaviors)
# ==========================================================
print("\nAll unique category names:")
print(X["category_name"].unique())

print("\nNumber of records per category:")
print(X["category_name"].value_counts())

keyword = "pizza"
print(f"\nRecords containing the keyword '{keyword}': (showing up to 5)")
print(X[X["text"].str.contains(keyword, case=False, na=False)][["text", "category_name"]].head(5))

print("\nRecords from index 100 to 110:")
print(X.iloc[100:111][["text", "category_name"]])

print("\nRecords in 'positive' category and containing 'stock': (up to 5)")
print(
    X[(X["category_name"] == "positive") & (X["text"].str.contains("stock", case=False, na=False))][["text", "category_name"]]
    .head(5)
)

# ==========================================================
# 3) Select every 10th record from a category, show first 5
# ==========================================================
cat = "positive"
cat_records = X[X["category_name"] == cat]
every_10th_record = cat_records.iloc[::10]
print(f"\nEvery 10th record in '{cat}' (first 5):")
print(every_10th_record.head(5)[["text", "category_name"]])

# ==========================================================
# 4) Missing values per record (row-wise) & NA example table
# ==========================================================
print("\nRow-wise missing counts (first 10 rows):")
print(X.isnull().sum(axis=1).head(10))

NA_dict = [
    {"id": "A", "missing_example": np.nan},
    {"id": "B"},
    {"id": "C", "missing_example": "NaN"},
    {"id": "D", "missing_example": "None"},
    {"id": "E", "missing_example": None},
    {"id": "F", "missing_example": ""},
]
NA_df = pd.DataFrame(NA_dict, columns=["id", "missing_example"])
print("\nNA example dataframe:")
print(NA_df)

# ==========================================================
# 5) Exercise 6 (take home) — Compare X vs X_sample diagnostics
# ==========================================================
print("\n### Differences between X and X_sample (diagnostics)")
print("Shape X vs X_sample:", X.shape, X_sample.shape)

full_counts = X["category_name"].value_counts().sort_index()
sample_counts = X_sample["category_name"].value_counts().sort_index()
diff_df = pd.DataFrame({"full": full_counts, "sample": sample_counts}).fillna(0).astype(int)
diff_df["sample_ratio"] = (diff_df["sample"] / diff_df["full"]).round(3)
print(diff_df)

# Compare average text length
X = X.assign(text_len=X["text"].str.len())
X_sample = X_sample.assign(text_len=X_sample["text"].str.len())
print("\nAvg text length — full vs sample:",
      round(X["text_len"].mean(), 2), "vs", round(X_sample["text_len"].mean(), 2))

# ==========================================================
# 6) Category distribution bar chart (auto ylim)
# ==========================================================
category_counts = X["category_name"].value_counts()
ymax = int(category_counts.max() * 1.1) + 1

plt.figure(figsize=(8, 3))
category_counts.plot(kind="bar", title="Category distribution", ylim=[0, ymax], rot=0, fontsize=11)
plt.xlabel("Category"); plt.ylabel("Count"); plt.tight_layout(); plt.show()

# ==========================================================
# 7) Compare category distributions: X vs X_sample
# ==========================================================
categories = sorted(list(set(X["category_name"]).union(set(X_sample["category_name"]))))
X_counts = X["category_name"].value_counts().reindex(categories, fill_value=0)
X_sample_counts = X_sample["category_name"].value_counts().reindex(categories, fill_value=0)

index = np.arange(len(categories)); bar_width = 0.35
plt.figure(figsize=(8, 4))
plt.bar(index, X_counts.values, bar_width, label="X full dataset")
plt.bar(index + bar_width, X_sample_counts.values, bar_width, label="X_sample sample dataset")
plt.title("Category Distribution Comparison: X vs X_sample")
plt.xlabel("Category"); plt.ylabel("Count")
plt.xticks(index + bar_width/2, categories)
plt.legend(); plt.tight_layout(); plt.show()

# ==========================================================
# 8) CountVectorizer analyzer & term-document matrix
# ==========================================================
count_vect = CountVectorizer(
    lowercase=True,
    stop_words=None,         # set to 'english' if desired
    max_features=50000,
    ngram_range=(1, 1)
)
X_counts = count_vect.fit_transform(X["text"])  # [n_docs, n_terms] sparse matrix
analyze = count_vect.build_analyzer()

print("\nTokenized first document:")
print(analyze(X.iloc[0]["text"]))

# Words that appear exactly once in the 5th document among first 100 features
if X_counts.shape[0] >= 5:
    vec_5 = X_counts[4, 0:min(100, X_counts.shape[1])].toarray()[0]
    word_indices = [i for i, v in enumerate(vec_5) if v == 1]
    words_once = count_vect.get_feature_names_out()[word_indices]
    print("\nWords that appear exactly once in the 5th document (first 100 features):")
    print(words_once)
else:
    print("\nDataset has fewer than 5 documents; skipping 'once in the 5th document' demo.")

# ==========================================================
# 9) Heatmap-like view: random 50x50 submatrix of counts (matplotlib only)
# ==========================================================
n_docs = min(50, X_counts.shape[0])
n_terms = min(50, X_counts.shape[1])
if n_docs > 0 and n_terms > 0:
    doc_idx = np.random.choice(X_counts.shape[0], n_docs, replace=False)
    term_idx = np.random.choice(X_counts.shape[1], n_terms, replace=False)
    sub = X_counts[doc_idx, :][:, term_idx].toarray()

    term_names = np.array(count_vect.get_feature_names_out())[term_idx]
    plt.figure(figsize=(10, 6))
    plt.imshow(sub, aspect="auto", interpolation="nearest")
    plt.title("Random 50x50 term-document heatmap (counts)")
    plt.xlabel("terms"); plt.ylabel("docs")
    plt.colorbar()
    plt.xticks(ticks=np.arange(n_terms), labels=term_names, rotation=90, fontsize=8)
    plt.yticks(ticks=np.arange(n_docs), labels=doc_idx, fontsize=8)
    plt.tight_layout(); plt.show()

# ==========================================================
# 10) Term frequencies (global), top-N bar plots (plain matplotlib)
# ==========================================================
term_frequencies = np.asarray(X_counts.sum(axis=0)).ravel()
term_frequencies_log = np.log1p(term_frequencies)
vocab = count_vect.get_feature_names_out()

# Top 50 by raw frequency
N = min(50, len(vocab))
top_idx = np.argsort(term_frequencies)[-N:][::-1]
top_terms = vocab[top_idx]; top_freqs = term_frequencies[top_idx]

plt.figure(figsize=(12, 4))
plt.bar(range(N), top_freqs)
plt.xticks(range(N), top_terms, rotation=90)
plt.title(f"Term Frequencies (Top {N})")
plt.xlabel("Terms"); plt.ylabel("Frequency")
plt.tight_layout(); plt.show()

# Top 300 by log frequency (efficient partial sort)
top_N = min(300, len(vocab))
approx_idx = np.argpartition(term_frequencies_log, -top_N)[-top_N:]
approx_idx_sorted = approx_idx[np.argsort(term_frequencies_log[approx_idx])[::-1]]
terms_300 = vocab[approx_idx_sorted]
logs_300 = term_frequencies_log[approx_idx_sorted]

plt.figure(figsize=(40, 5))
plt.bar(range(len(terms_300)), logs_300)
plt.xticks(range(len(terms_300)), terms_300, rotation=90)
plt.title("Top 300 Terms by Log Frequency")
plt.xlabel("Terms"); plt.ylabel("log(1+freq)")
plt.tight_layout(); plt.show()

# ==========================================================
# 11) Simple classifier like in master Phase 1 (Logistic Regression)
#     (optional but mirrors Exercises 10–15 style)
# ==========================================================
# For text, we'll use a very simple bag-of-words representation already built (X_counts)
# Split using the underlying numeric matrix
idx = np.arange(X_counts.shape[0])
train_idx, test_idx = train_test_split(idx, test_size=0.2, random_state=42, stratify=X["category_name"])
X_train_bow, X_test_bow = X_counts[train_idx], X_counts[test_idx]
y_train_cls = X.loc[train_idx, "category_name"].values
y_test_cls  = X.loc[test_idx, "category_name"].values

clf = LogisticRegression(max_iter=300)
clf.fit(X_train_bow, y_train_cls)
print("\nTrain acc:", round(clf.score(X_train_bow, y_train_cls), 4), 
        " Test acc:", round(clf.score(X_test_bow, y_test_cls), 4))

y_pred_cls = clf.predict(X_test_bow)
print("\nConfusion matrix (labels in order):", np.unique(y_train_cls))
print(confusion_matrix(y_test_cls, y_pred_cls, labels=np.unique(y_train_cls)))
print("\nClassification report:\n", classification_report(y_test_cls, y_pred_cls))

# Save and reload model (mirror Exercises 14–15)
joblib.dump(clf, "reddit_lr_bow.joblib")
print("\nReloaded model test acc:", joblib.load("reddit_lr_bow.joblib").score(X_test_bow, y_test_cls))

print("\n--- Phase 1 (Reddit-stock-sentiment) complete ---")
```

```python
### Begin Assignment Here

# DM2025 — Lab1 Phase 2 (NEW dataset only, no NB and no TF-IDF) — single cell
# Focus: frequent patterns, per-category top/bottom words, Count TDM, pattern augmentation, 3D visualization, reflections.

import sys, re, html, math, string, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from collections import Counter
from itertools import combinations
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

# ---------- Setup ----------
warnings.filterwarnings("ignore", category=FutureWarning)
pd.set_option("display.width", 140)
pd.set_option("display.max_colwidth", 140)
np.random.seed(42)
print("Python:", sys.version.split()[0])

# ---------- Load dataset ----------
URL = "https://raw.githubusercontent.com/leoson-wu/DM2025-Lab1-Exercise/main/newdataset/Reddit-stock-sentiment.csv"
df = pd.read_csv(URL)[["text","label"]].dropna()
df["label"] = df["label"].astype(float).clip(-1,1)
df = df[df["label"].isin([-1.0,0.0,1.0])].reset_index(drop=True)
label_map = {-1.0:"negative", 0.0:"neutral", 1.0:"positive"}
df["category_name"] = df["label"].map(label_map).astype("category")

def clean_text(s: str) -> str:
    s = str(s).lower()
    s = html.unescape(s)
    s = re.sub(r"http\S+|www\S+", " ", s)
    s = re.sub(r"[^a-z0-9$%\-\'\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

df["clean"] = df["text"].astype(str).map(clean_text)
print("Dataset shape:", df.shape)
print("Label counts:\n", df["category_name"].value_counts())

# ---------- Split ----------
X_train, X_test, y_train, y_test = train_test_split(
    df["clean"].values, df["category_name"].values,
    test_size=0.3, random_state=42, stratify=df["category_name"].values
)

# ---------- Frequent patterns (train only) ----------
def tokenize(s: str, min_len: int = 2):
    toks = s.split()
    out = []
    for t in toks:
        t = t.strip(string.punctuation)
        if len(t) >= min_len and not t.isdigit():
            out.append(t)
    return out

train_tokens = [tokenize(t) for t in X_train]

doc_freq = Counter()
for ts in train_tokens:
    doc_freq.update(set(ts))

MIN_SUP = 0.01
n_docs = len(train_tokens)
min_count = max(1, int(math.ceil(MIN_SUP * n_docs)))
keep_vocab = {w for w,c in doc_freq.items() if c >= min_count}
train_tokens = [[w for w in ts if w in keep_vocab] for ts in train_tokens]

item_counts = Counter()
for ts in train_tokens:
    item_counts.update(set(ts))
freq1 = [(w,c) for w,c in item_counts.items() if c >= min_count]
freq1.sort(key=lambda x:(-x[1], x[0]))

freq1_vocab = {w for w,_ in freq1}
pair_counts = Counter()
for ts in train_tokens:
    uniq = sorted(freq1_vocab.intersection(set(ts)))
    for a,b in combinations(uniq,2):
        pair_counts[(a,b)] += 1
freq2 = [((a,b),c) for (a,b),c in pair_counts.items() if c >= min_count]
freq2.sort(key=lambda x:(-x[1], x[0]))

TOP_K = 15
print(f"\nTop-{TOP_K} frequent 1-itemsets:")
for w,c in freq1[:TOP_K]:
    print(f"{w:>20s} : {c} (support={c/n_docs:.3f})")
print(f"\nTop-{TOP_K} frequent 2-itemsets:")
for (a,b),c in freq2[:TOP_K]:
    print(f"({a}, {b}) : {c} (support={c/n_docs:.3f})")

# ---------- Count TDM and per-category top/bottom words ----------
MIN_DF = 2
MAX_FEATURES = 40000
NGRAM = (1,1)
vec = CountVectorizer(min_df=MIN_DF, max_features=MAX_FEATURES, ngram_range=NGRAM)
X_counts = vec.fit_transform(df["clean"])
terms = np.array(vec.get_feature_names_out())

tdm_df = pd.DataFrame.sparse.from_spmatrix(X_counts, columns=terms)

for category in df["category_name"].cat.categories:
    mask = (df["category_name"].values == category)
    sub = X_counts[mask]
    sums = pd.Series(sub.sum(axis=0).A1, index=terms)
    sorted_words = sums.sort_values()
    total = len(sorted_words)
    top_n = max(1, int(0.05 * total))
    bottom_n = max(1, int(0.01 * total))
    print(f"\nCategory: {category}")
    print("Bottom 1% words:", list(sorted_words.head(bottom_n).index)[:15], "...")
    print("Top 5% words:", list(sorted_words.tail(top_n).index)[-15:], "...")

# ---------- Augment with frequent pair flags ----------
pair_list = [p for (p,_) in freq2]

def build_pair_flags(texts, pairs):
    pair_set = set(pairs)
    rows = []
    for t in texts:
        ts = set(tokenize(t))
        rows.append([1 if (a in ts and b in ts) else 0 for (a,b) in pair_set])
    return np.array(rows, dtype=np.int8)

pair_flags = build_pair_flags(df["clean"].tolist(), pair_list)
pattern_cols = [f"PAIR::{a}__{b}" for (a,b) in pair_list]
pattern_df = pd.DataFrame(pair_flags, columns=pattern_cols, dtype=np.int8)

augmented_df = pd.concat([tdm_df.sparse.to_dense(), pattern_df], axis=1)
print("\nAugmented feature shape:", augmented_df.shape)

# ---------- 3D visualization ----------
try:
    X_pca3 = PCA(n_components=3, random_state=42).fit_transform(augmented_df.values)
    X_tsne3 = TSNE(n_components=3, random_state=42, init="pca", learning_rate="auto").fit_transform(augmented_df.values)

    colors = ["coral", "blue", "black"]
    cats = df["category_name"].cat.categories.tolist()

    def plot_3d(ax, X3, title, elev, azim):
        for c, cat in zip(colors, cats):
            idx = (df["category_name"].values == cat)
            ax.scatter(X3[idx,0], X3[idx,1], X3[idx,2], c=c, label=cat, s=10, alpha=0.7)
        ax.set_title(title); ax.view_init(elev=elev, azim=azim)

    view_angles = [(20,30), (60,120)]
    fig = plt.figure(figsize=(12, 6))
    for j,(elev,azim) in enumerate(view_angles):
        ax = fig.add_subplot(1, 2, j+1, projection="3d")
        plot_3d(ax, X_tsne3, f"t-SNE 3D (elev={elev}, azim={azim})", elev, azim)
    plt.tight_layout(); plt.show()
except Exception as e:
    print("3D visualization skipped:", e)

# ---------- Reflections ----------
print("""
=== Reflections ===
Inefficiencies:
- Very rare tokens increase sparsity and noise.
- Pair mining can be slow with a large vocabulary.
- Simple tokenization ignores morphology and domain-specific tickers.

Improvements:
- Normalize tickers and keep finance markers ($, %).
- Use min_df to filter rare tokens and n-grams for collocations.
- Consider lemmatization/stemming if allowed.
- Restrict pair mining to top-N tokens for speed.
""")

print("\nDone — Phase 2 on NEW dataset (no NB, no TF-IDF).")
```
